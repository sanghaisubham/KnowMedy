```python
import csv
from collections import defaultdict

disease_list = []

def return_list(parameter):
    my_list = []
    match = parameter.replace('^','_').split('_')
    ctr = 1
    for group in match:
        if ctr%2==0:
            my_list.append(group)#Makes the list of disease or symptoms
        ctr = ctr + 1
    return my_list

with open("Unclean.csv") as csvfile:
    reader = csv.reader(csvfile)
    disease=""
    weight = 0
    disease_list = []
    dict_wt = {}
    #defaultdict returns new dictionary like object
    
    dict_=defaultdict(list) #Container implements specialized container datatypes 
                            #providing alternatives to Python's general purpose built-in containers,dict,list,
                            #set and tuple
    for row in reader:

        if row[0]!="\xc2\xa0" and row[0]!="":
            disease = row[0]
            disease_list = return_list(disease) #Make the list of disease

        if row[1]!="\xc2\xa0" and row[1]!="":
            symptom_list = return_list(row[1]) #Make the list of symptoms

            for d in disease_list:
                for s in symptom_list:
                    dict_[d].append(s)
    
```

```python
with open("cleaned_data.csv","w") as csvfile:
    writer = csv.writer(csvfile)
    for key,values in dict_.items():
        for v in values:
            #key = str.encode(key)
            key = str.encode(key).decode('utf-8')
            #.strip()
            #v = v.encode('utf-8').strip()
            #v = str.encode(v)
            writer.writerow([key,v])
```
