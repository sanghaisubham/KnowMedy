```python
dic={'fever':2,'cough':4}
```

```python
import pandas as pd
df=pd.read_csv('dataset_severe.csv',index_col=0)
```

```python
df
```

```python
Diseases=df['Source']
df=df.drop('Source',axis=1)
```

```python
col=df.columns
```

```python
col
```

```python
sym={}
for i in col:
    sym[i]=0
for k,v in dic.items():
    sym[k]=v
```

```python
sym
```

```python
df_test=pd.DataFrame([sym],columns=sym.keys())
```

```python
df_test
```

```python
from sklearn.externals import joblib
import pickle
classifier_model = joblib.load('bagging_classifier.pkl')
predicted_labels =classifier_model.predict(df_test)
```

```python
predict=classifier_model.predict_proba(df_test)
```

```python
predict
```

```python
predict.shape
```

```python
predict=pd.DataFrame(predict,columns=Diseases)
```

```python
predict
```

```python
Total=1
finalans={}
predict['heartburn']=0
#print(predict[predicted_labels[0]][0])
probab=predict[predicted_labels[0]][0]*3
finalans[0]=({'disease':predicted_labels[0],'prob':probab})
Total=Total-probab
# print(predict[predicted_labels[0]])
predict[predicted_labels[0]]=0
#finalans[append(predicted_labels[0])
for i in range(3):
    dis=predict.idxmax(axis=1)
    probab=predict[dis[0]][0]*3
    Total=Total-probab
    finalans[i+1]=({'disease':dis[0],'prob':probab})
    #finalans.append(dis[0])
    predict[predict.idxmax(axis=1)]=0
```

```python
predict.to_csv('pred_prob.csv')
```
